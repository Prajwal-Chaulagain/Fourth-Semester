```python
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide (a, b):
    if b != 0:
     return a / b
    else:
     return "Error: Division by zero"
 
def calculate(a, b, operation):
    if operation == '+':
        return add(a, b)
    elif operation == '-':
        return subtract(a, b)
    elif operation == '*':
        return multiply(a, b)
    elif operation == '/':
        return divide(a, b)
    elif operation == '^':
        return pow(a, b)
    else:
        return "Error: Invalid operation"
a = int(input("Enter a first number:"))
b = int(input("Enter a second number:"))
operation = input("Enter operation (+, -, *, /):")
result = calculate(a, b, operation)
print("Result:", result)
```

```python
def step_function(x):
    return 1 if x >= 0 else 0

# Generate training data
import random
def generate_data (n_samples=10):
    data = []
    labels = []
    for _ in range(n_samples):
        x = random.uniform(0, 10)
        decimal_part = x -int(x)
        label = 1 if decimal_part >= 0.5 else 0
        data.append([decimal_part])
        labels.append(label)
    return data, labels
    
generate_data()
```

```python
# Perceptron training
def train_perceptron(data, labels, learning_rate=0.1, epochs=10):
    weight = 0.0
    bias = 0.0
    
    for epoch in range(epochs):
        print(f"\nEpoch {epoch+1}")
        for i in range(len(data)):
            x = data[i][0]
            y_true = labels[i]
            z = x * weight + bias 
            y_pred = step_function(z)
            error = y_true - y_pred
        
            weight += learning_rate * error * x
            bias += learning_rate *error
        
            print(f"x: {x:.3f}, y: {y_true}, pred: {y_pred}, err: {error}, w: {weight:.3f}, b: {bias:.3f}")
    return weight, bias
    
train_perceptron([[0.1], [0.2], [0.3]], [1,1,1])

    
```

```python
# Test the trained perceptron
def test_perceptron(weight, bias):
    print("\nTesting on sample values:")
    for i in range(11):
        x = i + 0.3
        z = (x % 1) * weight + bias
        print(f"x= {x:.1f} => pred = {step_function(z)}")
        
    for i in range(5):
        x = i + 0.7
        z = (x % 1) * weight + bias   
        print(f"x= {x:.1f} => pred = {step_function(z)}")
        
data, labels = generate_data(100) 
weight, bias = train_perceptron(data, labels)
test_perceptron(weight, bias)       
```
